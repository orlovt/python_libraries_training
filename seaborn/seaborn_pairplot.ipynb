## sns.pairplot 
sns.pairplot is very useful for looking for relationships and correlations 

Some of the import arguments:

1. hue = df.col() where is column is categorical, this helps a lot when trying to determine correlations based on "type"

2. markers = [] sets markers for types in "HUE"

3. vars = [], makes y_ax and x_ax the same and plots pairplots 

4. y_vars and x_vars determine x and y axis, create corresponding plots

5. kind = 'reg' created a regression for every pairplot with different args 

6. diag_kind = 'kde' created a kernel estimated density line for the diagonal (if there is one )

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
sns.pairplot(iris, hue = 'species', markers = ['o', 's', 'd'])
```

```python
sns.pairplot(iris, hue = 'species', vars = ['sepal_width', 'sepal_length'], diag_kind = 'kde')
```

```python
sns.pairplot(iris, hue = 'species', x_vars = ['sepal_length', 'sepal_width'], y_vars = ['petal_length', 'petal_width'])
```

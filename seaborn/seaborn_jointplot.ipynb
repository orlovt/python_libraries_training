## sns.jointplot 

Provides distributions on the sides and a scatter plot in the middle

Some of the important arguments:

1. we set up data = data, x = data.col y = data.col

2. kind = 'reg' introduces linear regression with confedence boundaries and KDE curves for distributions 

3. kind = 'hex' outputs a plot simular to a heatmap 

4. kind = 'kde' produces a result which is simlular to kernel distribution estiate for a pair of variables 

5. we can also join kde plot on top of the scatterplot(middle) and a rugplot on top of the distplot(sides)

    
    
    (sns.jointplot("total_bill", "tip", data=tips)

            .plot_joint(sns.kdeplot, n_levels=5)

            .plot_marginals(sns.rugplot))

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip' )
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex') #like a heatmap 
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', shade = True)
```

```python
(sns.jointplot("total_bill", "tip", data=tips)
         .plot_joint(sns.kdeplot, n_levels=5)
         .plot_marginals(sns.rugplot))
```

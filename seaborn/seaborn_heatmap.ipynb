## sns.heatmap 
sns.heatmap is very useful for looking for relationships and correlations 

Some of the import arguments:

1. vmin and vmax sets the value for max corr for intercestions of params 

2. robust = True [2%, 98%] - cuts out the outliers 

3. we can pivot the data first and then display it with the heatmap

flights = flights.pivot('month', 'year', 'passengers') #agrs: x_ax, y_ax, value_counts!

4. annot = True shows numbers on top of the colored cells 

5. fmt = ''?''

link to different types of fmt https://docs.python.org/3/library/string.html?highlight=string#formatspec



```python
sns.heatmap(uniform_data)
```

```python
sns.heatmap(uniform_data, vmin = 0, vmax = 1, robust = True)
```

```python
normal_data = np.random.randn(10, 12)
```

```python
sns.heatmap(normal_data, robust = True)
```

## KDE 
KDE is mostly used only to show kernel estimated density function (PDF)

We can also call CDF function with sns.kdeplot(cumulative = True)

kdeplot can plot multiple (2) vectors of data (lookis great)

Other args:

1. shade = True, gives the plot a nice shade 

2. n_levels for 2 vectors of data gives n levels for compl distrs

3. gridsize determines the smooothness of the distplot 

4. cut determines how much to cut from the extreme datapoints (min/max)

5. clip determines the range of value for pdf 

```python
sns.kdeplot(tips.total_bill)
```

```python
sns.kdeplot(tips.total_bill, shade = True)
```

```python
sns.kdeplot(tips.total_bill, tips.tip, shade = True)
```

```python
sns.kdeplot(tips.total_bill, tips.tip, shade = True, n_levels = 10)
```

```python
sns.kdeplot(tips.total_bill, bw_method = 'silverman') 
```

```python
sns.kdeplot(tips.total_bill, shade = True, cumulative = True)
```

```python
sns.kdeplot(tips.total_bill, shade = True, gridsize = 20)
```

```python
sns.kdeplot(tips.total_bill, shade = True, cut = 0)
```

```python
sns.kdeplot(tips.total_bill, shade = True, clip = (5, 25))
```

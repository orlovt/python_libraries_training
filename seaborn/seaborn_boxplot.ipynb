## sns.boxplot 
sns.pairplot represents a box with quartiles and whiskers (almost outliers) and real outliers as points 

Some of the import arguments:

1. notch = True, creates a notch on the median (50% quartile)

2. we can also use two columns of data for these boxplots with sns.boxplot(x = 'day', y = 'total_bill', data = tips)

3. hue = df.col(categorical) makes a hue based on a categorical feature

We can also plot an sns.swarmplot on top of sns.boxplot 

```python
tips
```

```python
sns.boxplot(x = tips.total_bill)
```

```python
sns.boxplot(x = tips.total_bill, notch = True)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, notch = True)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, notch = True, hue = 'sex')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, c = '.25')
```
